```python
my_tasks = []
def add_task(task):
    my_tasks.append({"id": len(my_tasks) + 1, "task": task})
    print(f"Task added: '{task}'")
def view_tasks():
    print("YOUR TO-DO LIST:")
    print("-" * 30)
    if len(my_tasks) == 0:
        print("No tasks yet. Add task!")
    else:
        for index, item in enumerate(my_tasks, start=1):
            print(f"  {index}. {item['task']}")
    print("-" * 28
         )
def main():
    print("","=" * 28)
    print("  WELCOME TO TO-DO LIST APP")
    print("","=" * 28)

    while True:
        print(" MENU:")
        print("  1.Add Task")
        print("  2.View Tasks")
        print("  3.Exit")
        choice = input("Enter your choice 1/2/3: ")
        if choice == "1":
            task = input("Enter your task: ")
            if task:
                add_task(task)
            else:
                print("Task cannot be empty.")
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            print("Exit program..")
            break
        else:
            print("Invalid choice. Please enter valid choice 1, 2, or 3.")
if __name__ == "__main__":
    main()
```
